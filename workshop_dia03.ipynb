```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Criar DataFrame a partir dos dados
df = pd.read_csv('/content/dados03.csv')
```

```python
# Calcular valores mínimos, máximos e médios
pontuação_mínima = df[['math score', 'reading score', 'writing score']].min()
pontuação_máxima = df[['math score', 'reading score', 'writing score']].max()
pontuação_média = df[['math score', 'reading score', 'writing score']].mean()
```

```python
print("Valores Mínimos:")  # Imprime os valores mínimos
print(pontuação_mínima)
print("\nValores Máximos:") # Imprime os valores máximos
print(pontuação_máxima)
print("\nValores Médios:") # Imprime os valores médios
print(pontuação_média)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Agrupar por raça/etnia e calcular a média das notas de Matemática
math_scores_grouped = df.groupby('race/ethnicity')['math score'].mean()
```

```python
# Plotar gráfico de barras
plt.figure(figsize=(5, 6))
math_scores_grouped.plot(kind='bar', color='b')
plt.title('Nota Média em Matemática por Raça/Etnia')
plt.xlabel('Raça/Etnia')
plt.ylabel('Nota Média em Matemática')
plt.xticks(rotation=0)
plt.show()
```

```python
# Agrupar por nível de escolaridade dos pais e calcular a média das notas de Escrita
writing_scores_grouped = df.groupby('parental level of education')['writing score'].mean()
```

```python
# Plotar gráfico de barras
plt.figure(figsize=(8, 6))
writing_scores_grouped.plot(kind='bar', color='red')
plt.title('Nota Média em Escrita por Nível de Escolaridade dos Pais')
plt.xlabel('Nível de Escolaridade dos Pais')
plt.ylabel('Nota Média em Escrita')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Criar DataFrame a partir dos dados
df = pd.read_csv('/content/dados03.csv')
```

```python
# Agrupar por participação em curso preparatório e calcular a média das notas de Leitura
reading_scores_grouped = df.groupby('test preparation course')['reading score'].mean()
```

```python
# Plotar em um gráfico de barras
plt.figure(figsize=(5, 5))
reading_scores_grouped.plot(kind='bar', color='skyblue')
plt.title('Nota Média em Leitura por Grupo Racial/Etnia')
plt.xlabel('Grupo Racial/Etnia')
plt.ylabel('Nota Média em Leitura')
plt.xticks(rotation=0)
plt.show()
```
